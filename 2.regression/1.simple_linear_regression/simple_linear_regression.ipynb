```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load data
dataset  = pd.read_csv("Salary_Data.csv")
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# training simple linear regression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# predicting
y_pred = regressor.predict(X_test)
```

```python
# visualizing taining results
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```

```python
# visualizing test results
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, y_pred, color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```

```python
# making a single prediction
print(regressor.predict([[14]]))
```

In this case, the model predict that for an employee with 14 year of experience the salary is $ 157156,150

```python
# getting the final linear regression equation with the values of the coeficients
print(regressor.coef_)
print(regressor.intercept_)
```
